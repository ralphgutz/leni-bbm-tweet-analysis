# Scraping #LetLeniLead tweets from Oct. 4-9
This notebook scrapes tweets with hashtag #LetLeniLead from October 4 to 9. The scraping is divided into five batches with 50,000 maximum data each. Note that some tweets may not reflect here due to privacy restrictions of Twitter profiles. The datetime used is based on the Coordinated Universal Time (UTC).

## Batch 1 (Oct. 4-5)

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list1 = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper("#LetLeniLead since:2021-10-04 until:2021-10-05").get_items()):
    if i > 50000:
        break
    tweets_list1.append([tweet.date, tweet.content, tweet.lang, tweet.user.username, tweet.user.verified, tweet.user.created,
                         tweet.user.followersCount, tweet.user.friendsCount, tweet.mentionedUsers])
    
tweets_df1 = pd.DataFrame(tweets_list1, columns=["Datetime", "Text", "Language", "Username", "Verified",
                                                 "AccountCreated", "Followers", "Following", "MentionedUsers"])
```

```python
tweets_df1
```
